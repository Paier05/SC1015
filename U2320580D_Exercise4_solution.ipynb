### Problem 1 : Predicting SalePrice using GrLivArea
**Note : We observed during EDA that GrLivArea and SalePrice have a strong linear relationship with correlation 0.71.
In this problem, you will build a Linear Regression model to predict SalePrice using GrLivArea and judge its accuracy.<br>
a) Create appropriate datasets for Train and Test in an 80:20 ratio and fit a Linear Regression model on the Train
set to predict SalePrice using GrLivArea. Print the coefficients of your model and plot the regression line.**<br>

**b) Check the Goodness of Fit of the model on the Train set and Prediction Accuracy of the model on the Test set.
Print the metrics for Goodness of Fit and Prediction Accuracy that you think are appropriate in each scenario.**

```python
#import all the necessary library at the beginning
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
sb.set()
```

```python
#reading the file
house = pd.read_csv(r"C:\Users\paier\OneDrive\Desktop\SC1015 DSAI\Exercise 4\train.csv")
GrLivArea = pd.DataFrame(house['GrLivArea'])
SalePrice = pd.DataFrame(house['SalePrice'])
```

```python
#spilting the data set into train set and test set
GA_train, GA_test, SP_train, SP_test = train_test_split(GrLivArea, SalePrice, test_size = 0.2)
```

```python
#starting to train the model
linreg = LinearRegression()
linreg.fit(GA_train, SP_train)
```

```python
#printing the coefficient of the train data set
print("The intercept of the regression line is", linreg.intercept_)
print("The coefficient of the regression line is ",linreg.coef_)
```

```python
#plotting the regression line
SP_train_pred = linreg.predict(GA_train)
SP_test_pred = linreg.predict(GA_test)

#printing regression line for train data set
f = plt.figure(figsize = (16, 8))
plt.scatter(GA_train, SP_train)
plt.scatter(GA_train, SP_train_pred, color="r")
plt.show()

#printing regression line for test data set
f = plt.figure(figsize = (16, 8))
plt.scatter(GA_test, SP_test, color = "g")
plt.scatter(GA_test, SP_test_pred, color="y")
plt.show()
```

```python
#printing Goodness of Fit on the train data
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(GA_train, SP_train))
```

```python
#printing Predictoin Accuracy on the test data
print("Prediction Accuracy \tTest Dataset")
print("Mean Squared Error (MSE) \t:", mean_squared_error(SP_test, SP_test_pred))
```

I think it is better than the model obtained in Problem 1 because as the outliers of the dataset has been cleaned, it will have a  lower MSE compared to model in Problem 1 and almost the same Goodness of Fit of the model with the model in the problem 1. Thus, it will have a higher prediction accuracy compared to the model obtain in Problem 1.

```python
#making a new dataframe of GrLivArea and SalePrice
house_combine = pd.DataFrame(house[['GrLivArea', 'SalePrice']])

Q1 = [1129.5, 129975]
Q3 = [1776.75, 214000]

i = 0

outlier = [[],[]] # first list is GrLivArea and the second list is SalePrice

#finding the outliers and clean the outliers
for td in house_combine:
    IQR = Q3[i] - Q1[i]
    for data in house_combine[td]:
        if data < Q1[i] - 1.5*IQR or data > Q3[i] + 1.5*IQR:
            outlier[i].append(data)
            house_combine = house_combine.drop(house_combine[house_combine[td] == data].index)
    i = i + 1
    
#printing all the outliers    
print("The outlier of the GrLivArea are:")
for datas in outlier[0]:
    print(datas, end = ", ")
    
print()

print("The outlier of the SalePrice are:")
for datas in outlier[1]:
    print(datas, end = ", ")
```

```python
#making new model
new_GA = pd.DataFrame(house_combine['GrLivArea'])
new_SP = pd.DataFrame(house_combine['SalePrice'])

new_GA_train, new_GA_test, new_SP_train, new_SP_test = train_test_split(new_GA, new_SP, test_size = 0.2)

#start to train the model
linreg.fit(new_GA_train, new_SP_train)

#printing the coefficient of the model
print("The intercept of the regression line is", linreg.intercept_)
print("The coefficient of the regression line is ",linreg.coef_)

#plotting the regression line for the train data set
new_SP_train_pred = linreg.predict(new_GA_train)

f = plt.figure(figsize = (16, 8))
plt.scatter(new_GA_train, new_SP_train)
plt.scatter(new_GA_train, new_SP_train_pred, color="r")
plt.show()

#plotting the regression line for the test data set
new_SP_test_pred = linreg.predict(new_GA_test)
f = plt.figure(figsize = (16, 8))
plt.scatter(new_GA_test, new_SP_test, color = 'g')
plt.scatter(new_GA_test, new_SP_test_pred, color="y")
plt.show()
```

```python
#printing Goodness of Fit on train data set
print("Goodness of Fit of Model \tTrain Dataset")
print("Explained Variance (R^2) \t:", linreg.score(new_GA_train, new_SP_train))
```

```python
#printing Prediction Accuracy on the test data
print("Prediction Accuracy \tTest Dataset")
print("Mean Squared Error (MSE) \t:", mean_squared_error(new_SP_test, new_SP_test_pred))
```
